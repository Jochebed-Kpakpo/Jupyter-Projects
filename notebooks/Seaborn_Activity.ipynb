# Why Seaborn? Activity: Visualising Data with Heatmap, Boxplot, and Pairplot

This notebook introduces learners to Seaborn and its advantages over Matplotlib, focusing on creating and customising visualisations. 
You will work through three types of visualisations:
- **Heatmap**: To explore correlations between features.
- **Boxplot**: To compare distributions across categories.
- **Pairplot**: To analyse relationships between multiple variables.

## 1. Heatmap: Correlation Matrix
A heatmap is used to visualise correlations between features in a dataset. We'll use the Iris dataset for this demonstration.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Select only numeric columns
# The .corr() method can only compute correlations for numerical columns.
numeric_iris = iris.select_dtypes(include=['float', 'int'])

# Compute the correlation matrix
correlation_matrix = numeric_iris.corr()

# Create the heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Add title and axis labels
plt.title('Correlation Matrix', fontsize=14)
plt.xlabel('Features')
plt.ylabel('Features')
plt.show()
```

## 2. Boxplot: Compare Groups
A boxplot is used to compare distributions across categories. We'll visualise the distribution of sepal lengths across different species in Seaborn.

```python
# Boxplot to compare sepal length for different species
plt.figure(figsize=(8, 5))
sns.boxplot(
    x='species', 
    y='sepal_length', 
    data=iris, 
    palette='pastel', 
    hue='species',  # Explicitly set `hue` to match the x-axis variable
    dodge=False  # Prevent overlapping boxes
)

# Add labels and title
plt.title('Sepal Length by Species', fontsize=14)
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.show()
```

### Matplotlib version
Heres the Matplotlib version of the same boxplot for comparison:

```python
# Colors matching Seaborn's pastel palette
colors = ['#aec6cf', '#ffb347', '#b0e57c']  # Light blue, light orange, light green

# Boxplot using Matplotlib
plt.figure(figsize=(8, 5))
species = ['setosa', 'versicolor', 'virginica']
data = [iris[iris['species'] == sp]['sepal_length'] for sp in species]

# Create the boxplot with individual colors
box = plt.boxplot(data, patch_artist=True, widths=0.6)

# Apply Seaborn palette colors to each box
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Add labels and title
plt.title('Sepal Length by Species', fontsize=14)
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.xticks(ticks=[1, 2, 3], labels=species)

plt.show()
```

## 3. Pairplot: Relationships Between Variables
A pairplot is used to explore relationships between multiple variables. We'll use it to see patterns in the Iris dataset.

```python
# Pairplot to explore relationships between features
sns.pairplot(iris, hue='species', palette='husl')

# Add a title
plt.suptitle('Pairplot of Iris Features', y=1.02, fontsize=14)
plt.show()
```
