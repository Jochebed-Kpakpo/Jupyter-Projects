# Data Visualisation Challenges
This notebook contains both the original (incorrect) visualisations and improved visualisations.


## Scenario 1: Sales Trends Over 5 Years
### Original Visualisation: Bar Chart

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import random

data = {}

# Sales trends over 5 years (with monthly details)
data['years'] = list(range(2018, 2023))  # Using pure Python instead of np.arange
data['monthly_sales'] = {
    year: [500 + i * 50 for i in range(12)] for year in data['years']
}

# Original Visualisation: Bar Chart for Sales Trends
sales = [500, 600, 700, 800, 900]

plt.figure(figsize=(8, 5))
plt.bar(data['years'], sales, color='orange')
plt.title('Sales Trends Over 5 Years (Bar Chart)')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

### Improved Visualisation: Line Plot

```python
# Improved Visualisation: Line Plot for Sales Trends
plt.figure(figsize=(8, 5))
plt.plot(data['years'], sales, marker='o', linestyle='-', color='blue')
plt.title('Sales Trends Over 5 Years (Line Plot)')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

## Scenario 2: Comparison of Product Sales Across Regions
### Original Visualisation: Line Plot

```python
# Original Visualisation: Line Plot for Product Sales
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [200, 400, 300, 500]

plt.figure(figsize=(8, 5))
plt.plot(products, sales, marker='o', color='red')
plt.title('Comparison of Product Sales Across Regions (Line Plot)')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

### Improved Visualisation: Bar Chart

```python
# Improved Visualisation: Bar Chart for Product Sales
regions = ['North', 'South', 'East', 'West']
product_sales = {
    'Product A': [100, 150, 200, 250],
    'Product B': [300, 250, 200, 150],
    'Product C': [150, 200, 300, 350],
    'Product D': [400, 450, 500, 550]
}

df = pd.DataFrame(product_sales, index=regions)
df.plot(kind='bar', figsize=(10, 6))
plt.title('Comparison of Product Sales Across Regions (Bar Chart)')
plt.xlabel('Regions')
plt.ylabel('Sales')
plt.show()
```

## Scenario 3: Market Share Trends
### Original Visualisation: Scatter Plot

```python
# Market share over time
data['companies'] = ['Company A', 'Company B', 'Company C', 'Company D', 'Company E']
data['market_share'] = {
    company: [random.randint(30, 50) for _ in range(len(data['years']))] for company in data['companies']
}

plt.figure(figsize=(8, 5))
for company, market_share in data['market_share'].items():
    plt.scatter(data['years'], market_share, label=company)

plt.title('Market Share Trends (Scatter Plot)')
plt.xlabel('Years')
plt.ylabel('Market Share (%)')
plt.legend()
plt.show()
```

### Improved Visualisation: Line Plot

```python
import matplotlib.pyplot as plt
import random

# Data
data = {
    'years': [2018, 2019, 2020, 2021, 2022],
    'companies': ['Company A', 'Company B', 'Company C', 'Company D', 'Company E']
}
data['market_share'] = {
    company: [random.randint(30, 50) for _ in range(len(data['years']))] for company in data['companies']
}

# Plot Market Share Trends
plt.figure(figsize=(10, 6))
for company, market_share in data['market_share'].items():
    plt.plot(data['years'], market_share, marker='o', label=company)

# Chart Customisation
plt.title('Market Share Trends Over Time', fontsize=16)
plt.xlabel('Years', fontsize=12)
plt.ylabel('Market Share (%)', fontsize=12)
plt.grid(True)
plt.legend(title='Companies', fontsize=10)
plt.show()
```

## Scenario 4: Salary Distribution and Pay Gap Analysis
### Original Visualisation: Line Plot

```python
# Original Visualisation: Line Plot for Salaries
departments = ['HR', 'Engineering', 'Sales', 'Marketing']
average_salaries = [40000, 70000, 60000, 50000]

plt.figure(figsize=(8, 5))
plt.plot(departments, average_salaries, marker='o', linestyle='-', color='red')
plt.title('Salary Distribution and Pay Gap (Line Plot)')
plt.xlabel('Departments')
plt.ylabel('Average Salary')
plt.show()
```

### Improved Visualisation: Box Plot

```python
import random
import matplotlib.pyplot as plt

# Generate salaries using random.uniform to simulate normal-like distributions
def generate_salaries(mean, spread, count):
    return [random.uniform(mean - spread, mean + spread) for _ in range(count)]

# Data
salaries = {
    'HR': generate_salaries(40000, 5000, 100),
    'Engineering': generate_salaries(70000, 10000, 100),
    'Sales': generate_salaries(60000, 8000, 100),
    'Marketing': generate_salaries(50000, 7000, 100),
}

# Plot Box Plot
plt.figure(figsize=(8, 5))
plt.boxplot(salaries.values(), labels=salaries.keys())
plt.title('Salary Distribution and Pay Gap (Box Plot)')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```
